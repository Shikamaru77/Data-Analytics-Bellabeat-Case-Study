#### 5.2 Distribution of users based on their activity levels


According to the ['Counting Your Steps'](https://www.10000steps.org.au/articles/healthy-lifestyles/counting-steps/) article, step goal of 10,000 is the recommended daily step target for healthy adults. 

But based on the given dataset we can categorize the users into distinct groups of :
 - Sedentary level.
 - Lightly active level
 - Fairly active level
 - Very active level

```python
#calculating the minutes spent on levels
sedentary = merge_df['sedentaryminutes'].sum()
light_act = merge_df['lightlyactiveminutes'].sum()
fairly_act= merge_df['fairlyactiveminutes'].sum()
very_act = merge_df['veryactiveminutes'].sum()
total_time = sedentary+light_act+fairly_act+very_act

#percentage calculation
SEDperc = (sedentary/total_time)*100
LIGperc = (light_act/total_time)*100
FAIRperc = (fairly_act/total_time)*100
VERYperc = (very_act/total_time)*100

#taking the desired calculations in a list to plot for later reasons
activity_levels = [SEDperc,LIGperc,FAIRperc,VERYperc]
activity_labelling = ['Sedentary', 'Light Activity', 'Fairly Active', 'Very Active']
```

**PLOTTING**

```python
import matplotlib.pyplot as plt

# Plotting the pie chart
explode = [0.2, 0, 0, 0]
colors = ['#660000','#C51E3A','#FDBCB4','#29BF12']
plt.figure(figsize = (6,8))
plt.pie(activity_levels, labels= activity_labelling,colors = colors, wedgeprops = {"edgecolor": "black"}, 
        explode = explode, autopct = '%1.2f%%', textprops={'color': 'cyan'},startangle = 0 )

#annotations
plt.title('Activity Levels',fontweight='bold')

# Display the chart
plt.tight_layout()
plt.savefig('2-activityLVL.jpg')
plt.show()
```

The pie chart illustrates the distribution of user activity levels on the app as follows:
1. The majority of users, accounting for 81.33%, are not actively participating in daily activities or intensities.
2. Approximately 15.82% of users engage in light daily activities.
3. A small portion of users, comprising 1.74%, are classified as highly active, while 1.11% fall into the category of moderately active users.

In summary, the pie chart demonstrates that the majority of users are not actively engaged in daily activities, with a smaller percentage participating at different activity levels.

##### 5.3.1 Distribution of heart rates per weekday


Heart rate values can vary depending on factors such as age, fitness level, and overall health. Here are general guidelines for heart rate categories:

Resting Heart Rate:

1. Good: Typically, a resting heart rate **between 60 and 100** beats per minute (BPM) is considered normal for adults.
2. Average: The average resting heart rate for adults is **around 60-80 BPM**.
3. Bad: A resting heart rate consistently **above 100 BPM** or **below 60** BPM may indicate an underlying health condition and should be evaluated by a healthcare professional.
Target Heart Rate (during exercise):

****Good****: Target heart rate varies depending on the intensity of exercise, but it generally falls within a range of 50-85% of your maximum heart rate. This range ensures you're working at an appropriate level to benefit from exercise.

****Average****: The average target heart rate during exercise is around 70-85% of your maximum heart rate.

****Bad****: Exceeding your maximum recommended heart rate during exercise can put excessive strain on your cardiovascular system and may be considered bad. It's important to monitor your heart rate during exercise and stay within safe limits.

```python
#applying the above conditions and analysing

good = [val for val in df_sampledheartrate['value'] if 60 < val < 100]
avg = [val for val in df_sampledheartrate['value'] if 60 < val < 80]
bad = [val for val in df_sampledheartrate['value'] if val < 60 or val > 100]
```

```python
# Plotting the values in the 'good' list
sns.histplot(data=good, bins=4, color='green', label='Good')

sns.histplot(data=avg, bins=4, color='y', label='Average')

# Plotting the values in the 'bad' list
sns.histplot(data=bad, bins=12, color='red', label='Bad')

# Adding labels and legend
plt.title("Number of users' Heartrate(BPM)",fontweight = 'bold')
plt.xlabel('Heartrate(BPM)',fontweight = 'bold')
plt.ylabel('Count',fontweight = 'bold')
plt.legend()

plt.savefig('5-BPMfreq.jpg')

# Display the plot
plt.grid(True)
plt.show()
```

**Explanation:**

*SLEEP VS ACTIVITY LEVELS*

```python
#renaming all columns to small letter
df_sleep.head()
df_sleep.columns = df_sleep.columns.str.lower()

#column renaming
df_sleep.rename(columns = {'Id' : 'id'}, inplace = True)
df_sleep.rename(columns = {'sleepday' : 'activitydate'}, inplace = True)
df_sleep['activitydate'] = pd.to_datetime(df_sleep['activitydate'], format='%m/%d/%Y %I:%M:%S %p')

df_sleep.head()
```

```python
# Calculate the correlation matrix
correlation_matrix = merge_df[['totalminutesasleep', 'sedentaryminutes', 'lightlyactiveminutes', 'fairlyactiveminutes','veryactiveminutes']].corr()

#setting the figure size
plt.figure(figsize=(8, 6))

# Plot the correlation matrix as a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Customize the plot
plt.title('Correlation Between Sleep and Activity Levels', fontweight = 'bold')
plt.xlabel('Activity Levels', fontweight = 'bold')
plt.ylabel('Sleep', fontweight = 'bold')
plt.tight_layout()

plt.savefig('8-corrSLEEPvsACTlvl.jpg')

# Show the plot
plt.show()
```

**Explanation:**
   The above heatmap shows us the sleep vs activity levels correlation. It showed:
   1. The **sedentary minutes** showed a strong negative correlation against sleep providing us information that the increment of being idle causes low amount of sleep and vice-e-versa.Conversely, individuals who engage in less sedentary behavior tend to have longer sleep durations.

2. **Lightly active minutes**: There is a weak negative correlation of -0.045 between lightly active minutes and sleep. This suggests that there is a slight tendency for individuals who engage in more light physical activity to have slightly shorter sleep durations. However, the correlation is relatively weak, indicating that the relationship is not very strong.

3. **Fairly active minutes**: There is a moderate negative correlation of -0.140 between fairly active minutes and sleep. This implies that individuals who participate in more moderate levels of physical activity tend to have shorter sleep durations compared to those who engage in less moderate activity. The correlation coefficient indicates a relatively stronger relationship than with lightly active minutes.

4. **Very active minutes**: There is a weak negative correlation of -0.042 between very active minutes and sleep. This suggests that individuals who engage in high levels of physical activity have slightly shorter sleep durations. However, similar to lightly active minutes, the correlation is not very strong.

Overall, the correlations indicate that sedentary behavior has the strongest negative association with sleep duration, followed by fairly active minutes. Lightly active and very active minutes show weaker negative correlations, suggesting a less pronounced relationship with sleep duration.
